```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv("insurance.csv")
```

```python
df.head()
```

```python
plt.scatter(df.age,df.insurance)
```

```python
df.shape
```

```python
from sklearn.model_selection import train_test_split 
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(df[['age']],df.insurance,test_size=0.2,random_state=10)
```

```python
X_test
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
k=LogisticRegression()
```

```python
k.fit(X_train,Y_train)
```

```python
k.predict(X_test)
```

```python
X_test
```

```python
k.score(X_test,Y_test)
```
