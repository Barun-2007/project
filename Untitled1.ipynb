```python
import pandas as pd
import numpy as np
df=pd.read_csv("car.csv")
```

```python
dummy=pd.get_dummies(df.carmodel)
dummy
```

```python
merge=pd.concat([df,dummy],axis='columns')
merge
```

```python
k=merge.drop(['carmodel','Audi A5'],axis='columns')
k
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
```

```python
X=k.drop('mileage',axis='columns')
X
```

```python
Y=k.mileage
Y
```

```python
reg.fit(X,Y)
```

```python
reg.predict([[30000,5,False,False]])
```

```python
reg.score(X,Y)
```
